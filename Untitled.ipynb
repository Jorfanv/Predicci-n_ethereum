```python
import yfinance as yf
import matplotlib.pyplot as plt
import datetime
from matplotlib.animation import FuncAnimation

# Configura el símbolo para Ethereum en Yahoo Finance
symbol = 'ETH-USD'

# Inicializa el gráfico en tiempo real
fig, ax = plt.subplots()
x_data, y_data = [], []

def update(frame):
    current_time = datetime.datetime.now().strftime('%H:%M:%S')
    
    # Intenta obtener el precio, y maneja posibles errores
    try:
        price = yf.Ticker(symbol).info['regularMarketPrice']
    except:
        print("Error obteniendo el precio. Intentando de nuevo...")
        return

    x_data.append(current_time)
    y_data.append(price)

    ax.clear()
    ax.plot(x_data, y_data, label=f"{symbol} Price")
    ax.legend(loc="upper left")
    ax.set_title(f"{symbol} Price in Real-Time")
    ax.set_xlabel('Time')
    ax.set_ylabel('Price (USD)')
    ax.relim()  # Recalcula límites
    ax.autoscale_view()  # Ajusta los ejes
    plt.xticks(rotation=45)
    plt.tight_layout()

MAX_FRAMES = 100  # o el número que desees
ani = FuncAnimation(plt.gcf(), update, interval=60000, save_count=MAX_FRAMES)

plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime

# Configura el símbolo para Ethereum en Yahoo Finance
symbol = 'ETH-USD'
prices = []
timestamps = []

plt.ion()  # Activa el modo interactivo

fig, ax = plt.subplots()

def update_graph():
    # Obtiene el precio actual de Ethereum
    price = yf.Ticker(symbol).history(period="1d").tail(1)['Close'].iloc[0]
    current_time = datetime.now().strftime('%H:%M:%S')
    
    prices.append(price)
    timestamps.append(current_time)

    ax.clear()
    ax.plot(timestamps, prices, '-o', label=f"{symbol} Price")
    ax.set_title(f"{symbol} Price in Real-Time")
    ax.set_xlabel('Time')
    ax.set_ylabel('Price (USD)')
    plt.xticks(rotation=45)
    ax.legend(loc="upper left")
    plt.tight_layout()
    plt.draw()

# Actualiza el gráfico cada 60 segundos
while True:
    update_graph()
    plt.pause(60)
```

```python
import requests
import matplotlib.pyplot as plt
from datetime import datetime
import time

def get_ethereum_price():
    url = "https://api.coingecko.com/api/v3/simple/price?ids=ethereum&vs_currencies=usd"
    response = requests.get(url)
    return response.json()["ethereum"]["usd"]

def plot_ethereum_price(prices):
    plt.figure(figsize=(10, 5))
    plt.plot(prices, '-o')
    plt.title('Precio de Ethereum en tiempo real')
    plt.xlabel('Tiempo')
    plt.ylabel('Precio (USD)')
    plt.grid(True)
    plt.show(block=False)

def main():
    prices = []
    try:
        while True:
            price = get_ethereum_price()
            prices.append(price)
            print(f"Precio de Ethereum: {price} USD - {datetime.now().strftime('%H:%M:%S')}")
            plot_ethereum_price(prices)
            time.sleep(60)  # espera 1 minuto antes de obtener el próximo precio
            plt.close()

    except KeyboardInterrupt:
        plt.show()

if __name__ == "__main__":
    main()
```

```python
import requests
import matplotlib.pyplot as plt
from datetime import datetime
import time

def get_ethereum_price():
    url = "https://api.coingecko.com/api/v3/simple/price?ids=ethereum&vs_currencies=usd"
    response = requests.get(url)
    return response.json()["ethereum"]["usd"]

def main():
    plt.ion()  # Activa el modo interactivo
    fig, ax = plt.subplots(figsize=(10, 5))
    times = []
    prices = []

    try:
        while True:
            price = get_ethereum_price()
            current_time = datetime.now().strftime('%H:%M:%S')
            times.append(current_time)
            prices.append(price)
            
            if len(times) > 10:  # Limitar la cantidad de puntos en la gráfica para que sea legible
                times.pop(0)
                prices.pop(0)

            ax.clear()
            ax.plot(times, prices, '-o')
            ax.set_title('Precio de Ethereum en tiempo real')
            ax.set_xlabel('Tiempo')
            ax.set_ylabel('Precio (USD)')
            ax.grid(True)
            fig.canvas.draw()

            print(f"Precio de Ethereum: {price} USD - {current_time}")
            time.sleep(60)  # espera 1 minuto antes de obtener el próximo precio

    except KeyboardInterrupt:
        plt.ioff()  # Desactiva el modo interactivo
        plt.show()

if __name__ == "__main__":
    main()
```
